```python



from google.colab import drive

drive.flush_and_unmount()
print("Google Drive unmounted.")
import os
buildozerspec = "buildozer.spec"
if os.path.exists(buildozerspec):
    os.remove(buildozerspec)
    print("buildozer.spec deleted.")
else:
    print("buildozer.spec not found.")
# Celda 3: Definir los #@param para el proyecto
# @title Configuración del proyecto { run: "auto" }
# @markdown En esta celda podemos definir los parámetros que queremos usar para configurar nuestro proyecto con buildozer. Estos parámetros se guardarán en un archivo llamado buildozer.spec que buildozer usará para compilar el proyecto.

# Parámetros generales
app_title = "TftMixer" #@param {type:"string"}
app_domain = "tfs" #@param {type:"string"}
app_package = "fan.app.tft" #@param {type:"string"}
app_version = "1.0.0" #@param {type:"string"}
app_icon = "" #@param {type:"string"}
app_orientation = "portrait" #@param ["portrait", "landscape", "all"]
app_permissions = "" #@param {type:"string"}

# Parámetros de Python
python_version = "3.7" #@param ["2.7", "3.7", "3.11"]
requirements = "python3, sdl2_mixer,sdl2_ttf,png,jpeg,pygame,kivy,kivymd,jnius, sdl2, sdl2_image," #@param {type:"string"}
source_dir = "." #@param {type:"string"}
entry_point = "malisimo.py" #@param {type:"string"}

# Parámetros de Android
android_api = 25 #@param {type:"slider", min:21, max:29, step:1}
android_minapi = 24 #@param {type:"slider", min:21, max:29, step:1}
android_sdk = 25 #@param {type:"slider", min:19, max:29, step:1}
android_ndk = "25b" #@param ["25b","17c", "19b", "20b", "21d"]
android_arch = "arm64-v8a" #@param ["armeabi-v7a", "arm64-v8a", "x86", "x86_64"]
kivy_version = "2.3" #@param ["2.3", "2.2", "2.1", "2.0", "1.9"]
# Crear el archivo buildozer.spec con los parámetros definidos
!yes |buildozer init
#!runuser -l  allen -c 'buildozer init'
# Read existing buildozer.spec content
with open(buildozerspec, 'r') as spec_file:
    spec_content = spec_file.read()

# Replace or add necessary configurations
spec_content = spec_content.replace("title = My Application", f"title = {app_title}")
spec_content = spec_content.replace("package.name = myapp", f"package.name = {app_package}")
spec_content = spec_content.replace("version = 0.1", f"version = {app_version}")
spec_content = spec_content.replace("orientation = portrait", f"orientation = {app_orientation}")
spec_content = spec_content.replace("requirements = python3,kivy", f"requirements = {requirements}")
spec_content = spec_content.replace("source.dir = .", f"source.dir = {source_dir}")
#spec_content = spec_content.replace("#android.api = 31", f"android.api = {android_api}")
#spec_content = spec_content.replace("#android.minapi = 21", f"android.minapi = {android_minapi}")
#spec_content = spec_content.replace("#android.sdk = 20", f"android.sdk = {android_sdk}")
#spec_content = spec_content.replace("#android.ndk = 23b", f"android.ndk = {android_ndk}")
spec_content = spec_content.replace("android.archs = arm64-v8a, armeabi-v7a", f"android.arch = {android_arch}")# Write the modified content back to buildozer.spec
#spec_content = spec_content.replace("osx.kivy_version = 1.9", f"osx.kivy_version = {kivy_version}")
#spec_content = spec_content.replace("osx.python_version = 3", f"osx.python_version = {python_version}")
with open(buildozerspec, 'w') as spec_file:
    spec_file.write(spec_content)

print("buildozer.spec has been successfully modified.")
!yes |buildozer -v android debug
```
